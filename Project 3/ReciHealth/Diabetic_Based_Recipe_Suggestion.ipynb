##Installing dependencies

```python
import pickle
import numpy as np
import pandas as pd
```

##Load the dataset and trained models

```python
!git clone https://github.com/nithinrk11/Feynn-Labs-Machine-Learning-Internship.git
%cd Feynn-Labs-Machine-Learning-Internship
```

```python
# Load the saved model
from joblib import load
loaded_model = load('/content/Feynn-Labs-Machine-Learning-Internship/Project 3/trained_models/diabetes_model.pkl')
# Load the trained recipe prediction model from the pickle file
with open('/content/Feynn-Labs-Machine-Learning-Internship/Project 3/trained_models/trained_diab_recep_model.pkl', 'rb') as file:
    loaded_models = pickle.load(file)
# Load the dataset
data = pd.read_csv('/content/Feynn-Labs-Machine-Learning-Internship/Project 3/filtered_datasets/obese_recipe_df.csv')
```

```python
# Create a sample data point (replace with your data)
new_data = pd.DataFrame({
    'gender': [1],  # Male (1), Female (0)
    'age': [50],
    'hypertension': [1],  # Yes (1), No (0)
    'heart_disease': [0],  # Yes (1), No (0)
    'smoking_history': [1],  # Never (0), Former (1), Current (2), Not Current (3), ever (4), No Info (5)
    'bmi': [25],
    'HbA1c_level': [6],
    'blood_glucose_level': [100],
    'bmi_category':[1]
})



prediction = loaded_model.predict(new_data)

if prediction[0] == 1:
  diabetes_status = 'Diabetic'
else:
  diabetes_status = 'Non-Diabetic'

print(f'Prediction: {prediction}')
print(f'Diabetes Status: {diabetes_status}')
```

##Predict recipes and nutrional informations for diabetic friendly foods

```python
# Get the model for 'diabetic_friendly'
diabetic_model = loaded_models['diabetic_friendly']

# Make predictions for 'diabetic_friendly'
X_new = data[['Calories', 'Sugar (g)', 'Total Fat (g)', 'Saturated Fat (g)']]
diabetic_predictions = diabetic_model.predict(X_new)

# Filter the DataFrame for recipes predicted as diabetic_friendly
diabetic_friendly_recipes = data[diabetic_predictions == True]

# Select the desired columns (Food name, Cuisine, Sugar, Calories, Fat)
result_df = diabetic_friendly_recipes[['Recipe Name', 'Cuisine', 'Sugar (g)', 'Calories', 'Total Fat (g)']]

# Print the resulting DataFrame
result_df

```

Summary:
Most of the recipes in this list meet the diabetic-friendly criteria of low sugar (under 5g), calories (under 500), and fat (under 20g).
These recipes are diabetic-friendly, based on the nutritional information provided.

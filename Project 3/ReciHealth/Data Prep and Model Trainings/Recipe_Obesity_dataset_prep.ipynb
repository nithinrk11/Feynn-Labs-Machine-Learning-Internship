----
* Setting the nutrional citeria for different obesity types.
> The defined nutritional limits based on obesity risk types is an average estimation for more accurate and persoanlised levels, the prescribed levels by a proffesional is **prefered**

```python
# Define nutritional limits based on obesity risk types
criteria = {
    'Insufficient Weight': {
        'max_sugar': 30,
        'max_total_fat': 30,
        'max_saturated_fat': 15
    },
    'Normal Weight': {
        'max_sugar': 25,
        'max_total_fat': 25,
        'max_saturated_fat': 10
    },
    'Overweight_Level_I': {
        'max_sugar': 20,
        'max_total_fat': 20,
        'max_saturated_fat': 8
    },
    'Overweight_Level_II': {
        'max_sugar': 15,
        'max_total_fat': 15,
        'max_saturated_fat': 7
    },
    'Obesity_Type_I': {
        'max_sugar': 12,
        'max_total_fat': 12,
        'max_saturated_fat': 6
    },
    'Obesity_Type_II': {
        'max_sugar': 10,
        'max_total_fat': 10,
        'max_saturated_fat': 5
    },
    'Obesity_Type_III': {
        'max_sugar': 8,
        'max_total_fat': 8,
        'max_saturated_fat': 4
    }
}


# Add columns for each obesity risk type
for obesity_type, limits in criteria.items():
    recipe_df[f'{obesity_type}'] = (
        (recipe_df['Sugar (g)'] <= limits['max_sugar']) &
        (recipe_df['Total Fat (g)'] <= limits['max_total_fat']) &
        (recipe_df['Saturated Fat (g)'] <= limits['max_saturated_fat'])
    )

# Display the updated dataset
recipe_df.head()
```
